```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

import matplotlib.pyplot as plt

%matplotlib inline

X = np.linspace(-5, 9, num=100)[:, None]
y = -0.5 + 2.2*X +0.3*X**3+ 2*np.random.randn(100, 1)

X.shape
model = LinearRegression()
poly = PolynomialFeatures(degree=3, include_bias=False)
X_new = poly.fit_transform(X)
X_new.shape
model.fit(X_new[:70], y[:70])
y_pred = model.predict(X_new[5:])

model.fit(X[:55], y[:55])
y_pred1 = model.predict(X[5:])

fig,ax= plt.subplots(figsize=(15, 10))
ax.scatter(X, y,c = 'deeppink')
ax.set_facecolor('lightgray')
ax.plot(X[5:], y_pred, 'black',);
ax.plot(X[5:], y_pred1, c='orange',);
ax.legend(('Non-linear','Linear','training sample'), loc='upper center')
```
